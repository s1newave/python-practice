```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-8, 8, 0.01)
y = np.sin(2 * x) * np.cos(3 * x) / (x / 2)

plt.xlabel("Ось x")
plt.ylabel("Ось y")

yticks = np.arange(-2, 5)
plt.yticks(yticks, [f"{ytick:+}" for ytick in yticks])

plt.plot(x, y, color="r")

plt.show()
```
